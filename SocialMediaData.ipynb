```python
df['Interests'] = df['Interests'].apply(lambda x: x.replace("', ","',"))
```

```python
df
```

```python
interest_list = ','.join(list(df['Interests'])).split(',')
```

```python
interest_list = list(set(interest_list))
```

```python
for interest in interest_list:
    df[interest] = df['Interests'].apply(lambda x: 1 if interest in x else 0)
```

```python
df
```

```python
#or another way...
for interest in interest_list:
    df[interest] = np.where(df['Interests'].str.contains(interest), 1, 0)
```

```python
df_cluster = df.iloc[:, 9:df.shape[1]]  # Selecting columns from index 0 to 1 (inclusive)

```

```python
df.shape[1]
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3)  # Define the number of clusters
kmeans.fit(df_cluster)

# Get cluster labels for each sample
labels = kmeans.labels_

# Print cluster centers
print("Cluster Centers:")
print(kmeans.cluster_centers_)
```

```python
print("Cluster Labels:")
print(labels)
```
