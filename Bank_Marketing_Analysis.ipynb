```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

```python
# Create the Random Forest Regressor
rf_regressor = RandomForestRegressor(n_estimators=25, max_depth=10, random_state=42)

# Fit the model on the training data
rf_regressor.fit(df_nona[['log_income']], df_nona['Interest_Rate'])

# Predict on the test data
y_pred = rf_regressor.predict(df_nona[['log_income']])

# Evaluate the model
mse = mean_squared_error(df_nona['Interest_Rate'], y_pred)
print("Mean Squared Error:", mse)
# Calculate R-squared value
r_squared = r2_score(df_nona['Interest_Rate'], y_pred)
print("r squared:", r_squared)
```
