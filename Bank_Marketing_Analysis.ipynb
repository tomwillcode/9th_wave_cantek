Below is a great function that one CanTek student brought to my attention. basically, we can use curve_fit from scipy, to find the optimal paramaters based on least-squares, for whatever function we want to fit to the data, so below we don't know the optimal a, and b for that function, but scipy will find them and then fit it.

```python


def power_law(x, a, b):
    return a * np.power(x, b)

plt.scatter(df_nona['log_income'], df_nona['Interest_Rate'], alpha=0.7, label='Data')

x = df_nona['log_income']
y = df_nona['Interest_Rate']

params, covariance = curve_fit(power_law, x, y)
a, b = params

x_fit = np.linspace(min(x), max(x), 100)
y_fit = power_law(x_fit, a, b)
plt.plot(x_fit, y_fit, color='red', linestyle='-', linewidth=2, label='Power Law Fit')

plt.xlabel('Log of Income', fontsize=12)
plt.ylabel('Interest Rate', fontsize=12)
plt.title('Relationship Between Income and Interest Rate (Power Law)', fontsize=14)
plt.legend()

plt.grid(True)
plt.tight_layout()
plt.show()
```
