```python
import re

def contains_not_pattern(string):
    pattern = r'[^0-9/]'
    return re.search(pattern, string) is not None
```

```python
contains_not_pattern('12-')
```

```python
df['DOB'].apply(contains_not_pattern)
```

```python
df['DOB type'] = df['DOB'].apply(type)
```
